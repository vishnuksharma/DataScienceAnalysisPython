```python
# To convert the trending date column into date-time format, we need to explicilty define the pattern of the date

df['trending_date'] = pd.to_datetime(df['trending_date'], format = "%y.%d.%m")
```

```python
df.info()
```

```python
df.head(1)
```

```python
df['publish_time'] = pd.to_datetime(df['publish_time'].dt.date)
```

```python
df.info()
```

```python
df.head(1)
```

```python
df['days_took_to_trend'] = (df['trending_date'] - df['publish_time']).dt.days
```

```python
df.head(10)
```

```python
df['days_took_to_trend'].value_counts()
```

```python
df_hits = df[df['days_took_to_trend']==0]
df_hits.head()
```

```python
# If I don't mention any parameter within drop_duplicate, it will look for complete duplicate row
# It will drop the one and keep the first
df_hits.drop_duplicates(inplace = True, keep = 'first')
```

```python
# Within the videos that trended on the same day, which video is the most viewed video
df_hits.sort_values(by = "views", ascending = False)
```

```python
# Which channel has most number of same day trending videos
df_hits['channel_title'].value_counts()[:20]
```

```python
df_hits.groupby('channel_title').agg({"views":"sum"}).sort_values(by = "views", ascending = False)[:20]
```

```python
df_hits['category_id'].value_counts()
```
